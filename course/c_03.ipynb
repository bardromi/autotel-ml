### QQ Plot
* Used to determine whether a data set is distributed a certain way
* Usually showcases how the data fits a Normal Distribution

```python
scipy.stats.probplot(df_comp.spx,plot = pylab)
```

* X - How many standard deviations away from the mean these values are
* Red Line - what the data points should follow, if they are Normally Distributed

We can notice that the data isnt Normal distributed
